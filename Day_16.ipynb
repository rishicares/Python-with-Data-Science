```python
sns.lineplot(x = 'tip', y = 'day', data = tips)
```

```python
sns.scatterplot(x='total_bill', y = 'tip', hue = 'sex', data = tips)
```

```python
sns.scatterplot(x='total_bill', y = 'tip', hue = 'day', data = tips, palette='hot')
```

```python
#strip_plot
sns.stripplot(x='sex', y='total_bill', hue = 'day', data=tips)
```

```python
sns.swarmplot(x='sex', y = 'total_bill', hue='day', data=tips, palette = 'hot')
```

```python
# box plot
sns.boxplot(y='total_bill', x = 'sex', hue='day', data = tips, palette='hot')
```

```python
# violin plot
sns.violinplot(y='total_bill', x='sex', hue='day', data=tips, palette='hot', inner ='stick')
```
