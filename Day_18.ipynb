```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.boxplot(y='total_bill', x='day' ,data = tips, palette='hot')
```

```python
#despine()
sns.boxplot(y='total_bill', x='day' ,data = tips, palette='hot')
sns.despine()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(y='total_bill', x='sex', hue='day',data = tips, palette='hot')
sns.despine()
sns.set_context('paper', font_scale=1)
plt.grid()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(y='total_bill', x='sex', hue='day',data = tips, palette='hot')
sns.despine()
sns.set_context('paper', font_scale=1)
plt.grid()
plt.savefig('godiswatching.svg')
```
