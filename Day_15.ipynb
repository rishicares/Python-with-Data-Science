# Advance Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
```

```python
sine = np.sin(x)
cosine= np.cos(x)
tangent=np.tan(x)
exp = np.exp(x)

plt.plot(x,sine)
plt.plot(x,cosine)
plt.plot(x,tangent)
```

```python
fig, ax=plt.subplots(1,4, figsize=(30,5))
ax[0].plot(sine)
ax[1].plot(cosine)
ax[2].plot(tangent)
ax[3].plot(exp)
```

```python
fig, ax=plt.subplots(2,2, figsize=(20,10))
ax[0,0].plot(sine, label = 'Sine')
ax[0,0].plot(cosine, label='Cosine')
ax[0,0].set_title('Demo Sine Wave')
ax[0,0].legend()
ax[0,0].set_xlabel("Demo sine and cosine angles")
ax[0,0].set_ylabel("Demo sine and cosine values")

ax[0,1].plot(cosine)
ax[1,0].plot(tangent)
ax[1,1].plot(exp)
```
