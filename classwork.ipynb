```python
#creating a dataframe with columns a,b,c,d,e and values 1 to 25
```

```python
df=pd.DataFrame({"a":range(1,6),"b":range(6,11),"c":range(11,16),"d":range(16,21),"e":range(21,26)})
```

```python
df.head()
```

```python
df["e"]=df["a"]+df["b"]+df["c"]+df["d"]
df
```

```python
#............applying methods to the dataframe..............",it is necessary because we can't apply methods to the entire dataframe at once
def catagory(x):
    if x<10:
        return "0-10"
    elif x<20:
        return "10-20"
    elif x<30:
        return "20-30"
    else:
        return "30-40"
```

```python
df["catagory"]=df["e"].apply(catagory)
df
```

```python
#.............create a method to return square and appply it in the column c

def squre(x):
    return x**2
```

```python
df["square_of_c"]=df["c"].apply(squre)
df.head()
```

```python
#changing the order of the column
df=df[["a","b","c","d","e","catagory","square_of_c"]]
```

```python
df
```

```python
df.rename(columns={"a":"A","b":"B","c":"C","d":"D","e":"E"},inplace=True)
df
```

```python
columnorder=["A","B","C","square_of_c","D","E","catagory"]
df=df[columnorder]
df
```
