```python
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
```

```python
url='https://www.daraz.com.np/laptops/'
driver = webdriver.Chrome()
driver.get(url)
code=driver.page_source
```

```python
soup=BeautifulSoup(code, 'html.parser')
products=soup.find_all(class_='buTCk')
product_name=[]
product_price=[]
```

```python
for x in products:
    product_name.append(x.find(class_='RfADt').text)
    product_price.append(x.find(class_='ooOxS').text)
```

```python
product_price
```

```python
import pandas as pd
laptops=pd.DataFrame({'Product Name':product_name, 'Product Price':product_price})
laptops
```

```python
laptops.to_csv('daraz_laptops.csv')
```

```python
blog_url='https://www.setopati.com/social/338758'
driver = webdriver.Chrome()
driver.get(blog_url)
code=driver.page_source
```

```python
soup=BeautifulSoup(code, 'html.parser')
article=soup.find_all(class_='container')
title=soup.find(class_='news-big-title').text
content=[]
```

```python
for x in article:
    paragraphs = x.find_all('p')
    for p in paragraphs:
        content.append(p.text)
```
