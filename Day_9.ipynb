```python
# create function to categorize salary in to three different groups (low <=5000, middle <= 7000, and high)

def salary_group(salary):
    if salary <= 5000:
        return 'low'
    elif salary > 5000 and salary <= 7000:
        return'middle'
    else:
        return 'high'
er['category']= er['salary'].apply(salary_group)
```

```python
er
```

```python
#number of married people having high salary
cd1 = er['relationship']=='Married'
cd2 = er['category']=='high'

er[cd1 & cd2].value_counts()


# er[(er['relationship']=='Married') & (er['category']=='high')].value_counts()
```

```python
def ac(age):
    if age <= 30:
        return 'young'
    elif age > 30 and age <= 40:
        return 'middle'
    else:
        return 'old'

er['age_category'] = er['age'].apply(ac)

er
```
