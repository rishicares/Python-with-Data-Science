# Data Encoding

```python
#for machine learning
#nominal - non logic
#ordinal - logic
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.DataFrame({"name":["Prabal","Rishikesh","Anju","Prakriti","Manisha"],"size":["medium","large","small","small","medium"]})
```

```python
encoder=LabelEncoder()
```

```python
df["encoded"]=encoder.fit_transform(df["size"])
```

```python
df
```

```python
#ordinal encoding
size_mapping={"small":1,"medium":2,"large":3}
df["new encode"]=df["size"].map(size_mapping)
```

```python
df
```

```python
#............frequency encoding...............
```

```python
data=pd.DataFrame({'Category':['a','b','c','a','b','d','e','a','b','c','f']})
data
```

```python
frequency_encoding=data["Category"].value_counts()
data["Category encoded"]=data["Category"].map(frequency_encoding)
data
```
