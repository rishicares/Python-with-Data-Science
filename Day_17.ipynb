```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# Pair Plot
sns.pairplot(vars=['tip','total_bill','size'], diag_kind= 'auto', data = tips, hue='day')
```

```python
#Seaborn Distribution Plots
```

```python
sns.histplot(x='tip', kde=True, data = tips, hue='day', palette='hot')
```

```python
# kde(kernel density estimator)
```

```python
sns.kdeplot(x='total_bill', data = tips, hue='day')
```

```python
# Joint Plot
sns.jointplot(x='total_bill', y = 'tip', data = tips, palette='hot', hue='day')
```

```python
# rug plot
sns.rugplot(x='total_bill', data=tips, height=0.5, hue='day')
```

```python
# Creating a box plot
plt.figure(figsize=(20,10))
sns.boxplot(x='sex', y = 'total_bill', hue='day', data=tips, linewidth=0.5, palette = 'hot')
sns.set_context('poster', font_scale=1)
sns.despine()
```
